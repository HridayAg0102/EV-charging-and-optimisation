```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("minor.csv")
```

```python
df.head()
```

```python
# Power ratings in kW
power_ratings = {
    "PC":0.1,
    "TV":0.15,
    "Water Pump": 0.75,
    "Vacuum Cleaner":0.74,
    "Iron Box": 1.1,
    "Dryer":5.5,
    "Coffee Maker": 0.35,
    "Range Top (S)": 1.6,
    "Range Top (L)": 2.7,
    "Microwave Oven": 0.8,
    "Toaster": 1.1,
    "Toaster Oven":1.5,
    "Oven Cleaner": 3.5,
    "Washing Machine": 0.665,
    "Dish Washer": 1.2,
    "Oven": 3.5,
    "PHEV": 3.3,
    "AC": 1,
    "Fridge": 0.145,
    "Light": 0.16,
    "Fan": 0.06
}
```

```python
# This function finds the number of units consumed in a 15 minutes slot
# All appliances which were ON during the slot, are marked by 1, else 0.
# he "units_of_this_row" variable is a multiple of 0.25 to denote (15 min / 60 min) hour

def find_power_units_consumed(df, power_ratings):
    units = []
    for i in df.index:
        units_of_this_row = 0
        for j in df.columns[1:]:
            if df[j][i] == 1:
                units_of_this_row += 0.25*power_ratings[j]
        units.append(units_of_this_row)
        
    return units
        
```

```python
units = find_power_units_consumed(df, power_ratings)
```

```python
units = [round(units[i], 2) for i in range(len(units))]
```

```python
Units = pd.DataFrame(
{
    "Units_Consumed":units
})
```
